```python
import Perceptron
import sklearn.datasets
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split 
import matplotlib.pyplot as plt
```

```python
perceptron = Perceptron.Perceptron()
```

```python
breast_cancer =  sklearn.datasets.load_breast_cancer()
x = breast_cancer.data
y = breast_cancer.target
data = pd.DataFrame(x, columns=breast_cancer.feature_names)
data['class'] = y
x = data.drop('class', axis = 1)
y = data['class']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=1, stratify=y)
# Binarisation is not required in Perceptron
#x_train = x_train.apply(pd.cut, bins=2, labels = [1, 0])
#x_test = x_test.apply(pd.cut, bins=2, labels=[1, 0])
x_train = x_train.values
x_test = x_test.values
```

```python
perceptron.fit(x_train, y_train, 10, 0.01)
```

```python
perceptron.test_accuracy(x_test, y_test)
```

# Animation

```python
%matplotlib inline
from matplotlib import animation, rc
from IPython.display import HTML
```

```python
wt_matrix =  perceptron.fit(x_train, y_train, 100, 1, True)
print(wt_matrix.shape)
```

```python
# First set up the figure, the axis, and the plot element we want to animate
fig, ax = plt.subplots()

ax.set_xlim(( 0, wt_matrix.shape[1]))
ax.set_ylim((-12000, 26000))

line, = ax.plot([], [], lw=2)
```

```python
# animation function. This is called sequentially
def animate(i):
    x = list(range(wt_matrix.shape[1]))
    y = wt_matrix[i, :]
    line.set_data(x, y)
    return (line,)
```

```python
# call the animator. blit=True means only re-draw the parts that have changed.
anim = animation.FuncAnimation(fig, animate, frames=100, interval=20, blit=True)
```

```python
HTML(anim.to_html5_video())
```
