```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
x = np.linspace(0, 100, 1000)
```

```python
y = np.power(x, 0.33)
```

```python
sns.set()
sns.set_style('darkgrid')
sns.lineplot(x, y)
plt.show()
```

```python
plt.plot(x, y '*')
plt.show()
```
