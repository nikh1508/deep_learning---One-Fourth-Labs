```python
def plot_vectors(vec):
    colors = ['red', 'green', 'yellow', 'purple']
    ctr = 0
    for i in vec:
        plt.quiver(i[0], i[1], i[2], i[3], scale_units='xy', angles='xy', scale=1, color=colors[ctr])
        ctr += 1
    plt.xlim(-10, 10)
    plt.ylim(-10, 10)
    plt.show()
```
